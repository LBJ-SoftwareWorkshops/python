```python
import pandas as pd
salaries = pd.read_csv("ds_salaries.csv") 
salaries
```

```python
### cut

A = pd.Series([5, 0, 2, 8, 4, 10, 6, 7, 8, 2, 1, 8])
A

bins_A = pd.cut(A, bins = 4)
bins_A
bins_A.value_counts()
```

```python
### cut

A = pd.Series([75, 78, 80, 90, 99, 1, 3, 2, 45, 67])
A

bins_A = pd.cut(A, bins = 4)
bins_A
bins_A.value_counts()
```

```python
### qcut

B = pd.Series([75, 78, 80, 90, 99, 1, 3, 2, 45, 67])
bins_B = pd.qcut(B, q = 4)
bins_B.value_counts()
```
