```python
selected_columns = ["work_year", "job_title", "employment_type", "salary"]
salaries[selected_columns]

# ^ this, is the same as this:

salaries[["work_year", "job_title", "employment_type", "salary"]]
```

```python
full_time = salaries[salaries.employment_type == "FT"]
full_time2 = salaries[salaries["employment_type"] == "FT"]
full_time.head()

### The first method can only be used, when the variable name doesn't have a space
### Spaces in methods confuse python (pandas)
```

```python
### Filtering by multiple conditions:

full_time2020 = salaries[(salaries["work_year"] == 2020) | (salaries["work_year"] == 2021)]

### ^ is the same as

years2020_2021 = salaries[salaries["work_year"].isin([2020, 2021])]

#years2020_2021

### When you want to filter by NOT meeting the condition:

not2020_2021 = salaries[~salaries["work_year"].isin([2020, 2021])]
not2020_2021
```

```python
before_2021 = salaries.query("work_year < 2021 and job_title == 'Data Scientist' and employment_type == 'FT' and company_location == 'US'")
before_2021
```

```python
### Basic groupby function with mean
salaries[["job_title", "salary"]].groupby("job_title").mean()
```

```python
### groupby function/method with aggregate columns

aggregated = salaries.groupby(
        ["experience_level", "employment_type"]
    ).agg(
        average_salary = ("salary", "mean"),
        total_salary = ("salary", "sum")
            )
aggregated
```
